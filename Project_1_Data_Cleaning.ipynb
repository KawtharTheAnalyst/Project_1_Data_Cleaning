Creating a copy of the dataset to preserve the original data.

```python
df.copy() 
healthcare = df.copy()
```

Identifying and counting duplicate records in the dataset.

```python
healthcare.duplicated(subset = 'Patient_ID').sum()
```

Removing duplicate records while retaining the first occurence of each.

```python
healthcare = healthcare.drop_duplicates(subset = 'Patient_ID' , keep='first')
```

Comfirming that all duplicate records have been successfully removed.

```python
healthcare.duplicated(subset='Patient_ID').sum()
```

Checking the format and values of the Admission Date column.

```python
healthcare['Admission_Date'].value_counts().head(10)
```

Converting all Admission Date values to a uniform YYYY-MM-DD format.

```python
from datetime import datetime 
def fix_date(val):
    for fmt in ("%Y-%m-%d","%m/%d/%Y","%d-%m-%Y"):
        try:
            return datetime.strptime(str(val),fmt).strftime("%Y-%m-%d")
        except:
            continue
    return None
healthcare['Admission_Date'] = healthcare['Admission_Date'].apply(fix_date)
```

Confirming all dates are now in the correct YYYY-MM-DD format.

```python
healthcare['Admission_Date'].value_counts().head(10)
```

Checking the unique values and distribution of the Gender column.

```python
healthcare['Gender'].value_counts()
```

Replacing inconsistent Gender entries with uniform Male/Female values.

```python
healthcare['Gender'] = healthcare['Gender'].replace({
    'Female': 'Female',
    'female': 'Female',
    'F':'Female',
    'Male':'Male',
    'male':'Male',
    'M':'Male'})
```

Confirming all Gender values are now consistent.

```python
healthcare['Gender'].value_counts()
```

Veryfying the data types of all columns after cleaning.

```python
healthcare.dtypes
```

Converting the Admission Date column to the correct datetype(datetime).

```python
healthcare['Admission_Date'] = pd.to_datetime(healthcare['Admission_Date'])
```

Changing Admission Type data type to category.

```python
healthcare['Admission_Type'] = healthcare['Admission_Type'].astype('category')
```

Checking for inconsistent values in the Smoking Status column.

```python
healthcare['Smoking_Status'].value_counts()
```

Replacing inconsistent Smoking Status entries with uniform values.

```python
healthcare['Smoking_Status'] = healthcare['Smoking_Status'].replace({
    'Non-Smoker':'Non-Smoker',
    'Former Smoker':'Former-Smoker',
    'Smoker':'Smoker',
    'Never':'Never',
    'non-smoker':'Non-Smoker',
    'Smoker':'Smoker',
    'former smoker':'Former-Smoker',
    'never':'Never',
    'smoker':'Smoker'})
```

Standardizing text case in Diagnosis and Admission_Type columns

```python
healthcare['Diagnosis'] = healthcare['Diagnosis'].str.strip().str.title()
healthcare['Admission_Type'] = healthcare['Admission_Type'].str.strip().str.title()
```

Confirming all Smoking Status values are now consistent.

```python
healthcare['Smoking_Status'].value_counts()
```

Filling missing Length of Stay values using the median.

```python
healthcare['Length_of_Stay_Days'] = healthcare['Length_of_Stay_Days'].fillna(healthcare['Length_of_Stay_Days']).median()
```

Filling missing Doctor ID values with Unknown

```python
healthcare['Doctor_ID'] = healthcare['Doctor_ID'].fillna('Unknown')
```

Converting negative values to positive using abs() and ensuring correct data type with astype().

```python
healthcare['Length_of_Stay_Days'] = healthcare['Length_of_Stay_Days'].abs().astype(int)
```

Reviewing the cleaned dataset to confirm all changes were successfully applied.

```python
healthcare.info()
```

Filling missing categorical columns with mode and numerical columns with median.

```python
cat_cols = ['Gender','Diagnosis','Medication',
            'Smoking_Status','Insurance_Type',
            'Hospital','Readmission','Outcome']
for col in cat_cols:
    healthcare[col] = healthcare[col].fillna(healthcare[col].mode()[0]).astype('category')
```

```python
num_cols = ['BMI','Systolic_BP','Diastolic_BP',
            'Glucose_Level','Cholesterol','Hemoglobin']
for col in num_cols:
    healthcare[col] = healthcare[col].fillna(healthcare[col].median())
```

Confirming all missing values have been successfully handled.

```python
healthcare.info()
```

Filling missing Admission Type values with the most frequent value(mode).

```python
healthcare['Admission_Type'] = healthcare['Admission_Type'].fillna(healthcare['Admission_Type'].mode()[0])
```

Filling missing Doctor ID values with the most frequent value(mode).

```python
healthcare['Doctor_ID'].fillna(healthcare['Doctor_ID'].mode()[0])
```

Filling missing Admission Date values with the most frequent value(mode).

```python
healthcare['Admission_Date'] = healthcare['Admission_Date'].fillna(healthcare['Admission_Date'].mode()[0])
```

Confirming all columns now have 9305 non-null vahlues with no missing data remaining.

```python
healthcare.info()
```

Confirming all columns have the correct data types after cleaning.

Reviewing the descriptive statistics of all numerical columns after cleaning.

```python
healthcare.describe()
```

Verifying there are no negative Systolic BP values or extreme Glucose Level values above 500.

```python
print(healthcare[healthcare['Systolic_BP'] < 0].shape[0])
print(healthcare[healthcare['Glucose_Level'] > 500].shape[0])
```

Replacing negative, zero and unrealistic values(999) in Systolic BP with the median to ensure data integrity.

```python
median_sbp = healthcare[healthcare['Systolic_BP'] > 0]['Systolic_BP'].median()
healthcare.loc[healthcare['Systolic_BP'] < 0 ,'Systolic_BP'] = median_sbp
healthcare.loc[healthcare['Systolic_BP'] == 0,'Systolic_BP'] = median_sbp
healthcare.loc[healthcare['Systolic_BP'] == 999,'Systolic_BP'] = median_sbp
```

Confirming minimum and maximum Systolic BP values are now within realistic range.

```python
print('Min Systolic BP:', healthcare['Systolic_BP'].min())
print('Max Systolic BP:', healthcare['Systolic_BP'].max())
```
