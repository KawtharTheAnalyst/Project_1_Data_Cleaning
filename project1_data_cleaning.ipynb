```python
from datetime import datetime 
def fix_date(val):
    for fmt in ("%Y-%m-%d","%m/%d/%Y","%d-%m-%Y"):
        try:
            return datetime.strptime(str(val),fmt).strftime("%Y-%m-%d")
        except:
            continue
    return None
healthcare['Admission_Date'] = healthcare['Admission_Date'].apply(fix_date)
```

```python
healthcare['Admission_Date'].value_counts().head(10)
```

```python
healthcare['Gender'].value_counts()
```

```python
healthcare['Gender'] = healthcare['Gender'].replace({
    'Female': 'Female',
    'female': 'Female',
    'F':'Female',
    'Male':'Male',
    'male':'Male',
    'M':'Male'})
```

```python
healthcare['Gender'].value_counts()
```

```python
healthcare.dtypes
```

```python
healthcare['Admission_Date'] = pd.to_datetime(healthcare['Admission_Date'])
```

```python
healthcare['Admission_Type'] = healthcare['Admission_Type'].astype('category')
```

```python
healthcare['Smoking_Status'].value_counts()
```

```python
healthcare['Smoking_Status'] = healthcare['Smoking_Status'].replace({
    'Non-Smoker':'Non-Smoker',
    'Former Smoker':'Former-Smoker',
    'Smoker':'Smoker',
    'Never':'Never',
    'non-smoker':'Non-Smoker',
    'Smoker':'Smoker',
    'former smoker':'Former-Smoker',
    'never':'Never',
    'smoker':'Smoker'})
```

```python
healthcare['Smoking_Status'].value_counts()
```

```python
healthcare['Length_of_Stay_Days'] = healthcare['Length_of_Stay_Days'].fillna(healthcare['Length_of_Stay_Days']).median()
```

```python
healthcare['Length_of_Stay_Days'] = healthcare['Length_of_Stay_Days'].abs().astype(int)
```

```python
healthcare.info()
```

```python
cat_cols = ['Gender','Diagnosis','Medication',
            'Smoking_Status','Insurance_Type',
            'Hospital','Readmission','Outcome']
for col in cat_cols:
    healthcare[col] = healthcare[col].fillna(healthcare[col].mode()[0]).astype('category')
```

```python
num_cols = ['BMI','Systolic_BP','Diastolic_BP',
            'Glucose_Level','Cholesterol','Hemoglobin']
for col in num_cols:
    healthcare[col] = healthcare[col].fillna(healthcare[col].median())
```

```python
healthcare.info()
```

```python
healthcare['Admission_Type'] = healthcare['Admission_Type'].fillna(healthcare['Admission_Type'].mode()[0])
```

```python
healthcare['Doctor_ID'].fillna(healthcare['Doctor_ID'].mode()[0],inplace=True)
```

```python
healthcare['Admission_Date'] = healthcare['Admission_Date'].fillna(healthcare['Admission_Date'].mode()[0])
```
